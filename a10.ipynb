```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.stem import PorterStemmer,WordNetLemmatizer
```

```python
docs=[]
for i in range (1,11):
    with open(f"{i}.txt",'r') as file:
        docs.append(file.read())

docs
```

```python
import string
def Tokenize(doc):
    punctuation=string.punctuation+'[]<>{}'
    for char in punctuation:
        doc=doc.replace(char," ")
    doc=doc.lower()
    tokens=doc.split()
    return tokens
    
all_tokens=[]
for doc in docs:
    tokens=Tokenize(doc)
    all_tokens.append(tokens)

all_tokens
```

```python
def removeStopWords(tokens):
    stop_words=set(stopwords.words('english'))
    filtered_tokens=[]
    for token in tokens:
        if token not in stop_words:
            filtered_tokens.append(token)
    return filtered_tokens

pure_tokens=[]
for tokens in all_tokens:
    pure_tokens.append(removeStopWords(tokens))
pure_tokens
```

```python
posclass=pos_tag(pure_tokens[0])
posclass
```

```python
stemmer=PorterStemmer()
for token in pure_tokens[0]:
    stemmed_word=stemmer.stem(token)
    print(f"original word : {token} , stemmed word : {stemmed_word}")
```

```python
lemmatizer=WordNetLemmatizer()
for token in pure_tokens[0]:
    stemmed_word=lemmatizer.lemmatize(token)
    print(f"original word : {token} , lemmatized word word : {stemmed_word}")
```

```python
def calculateTF(tokens):
    n=len(tokens)
    tf_dict={}
    for token in tokens:
        if token not in tf_dict:
            tf_dict[token]=tokens.count(token)/n
    return tf_dict
    
all_tf=[]
for tokens in pure_tokens:
    all_tf.append(calculateTF(tokens))

all_tf
```

```python
import math
def calculateIDF(docs):
    n=len(docs)
    uniq_words=set()
    all_tokens=[]
    for doc in docs:
        tokens=word_tokenize(doc)
        all_tokens.append(tokens)
        for token in tokens:
            uniq_words.add(token)

    all_idf={}
    for word in uniq_words:
        cnt=0
        for tokens in all_tokens:
            if word in tokens:
                cnt+=1
        all_idf[word]=math.log(n/cnt)
    return all_idf

all_idf=calculateIDF(docs)
print(all_idf)
```

```python
all_tf_idf = []
i=0
for  tokens in all_tf:
    single_doc_tfidf = {}
    for word in tokens:
        if word in (all_tf[i] and all_idf):
            single_doc_tfidf[word] = all_tf[i][word] * all_idf[word]
    i+=1
    all_tf_idf.append(single_doc_tfidf)

print(all_tf_idf)

```
