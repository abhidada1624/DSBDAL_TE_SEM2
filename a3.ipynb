```python
def minVal(items):
    mini=100;
    for item in items:
        if mini>item:
            mini=item
    return mini
```

```python

def maxVal(items):
    maxi=0;
    for item in items:
        if maxi<item:
            maxi=item
    return maxi

```

```python
def giveMedian(data):
    items=sorted(data)
    n=len(data)
    val=0
    if n%2==0:
        val=(items[int(n/2)]+items[int(n/2)-1])/2
    else :
        val=items[int(n/2)]
    return val
```

```python

def statistic(data):
    mean=sum(data)/len(data)
    median=giveMedian(data)
    minval=minVal(data)
    maxval=maxVal(data)
    var=sum((x-mean)**2 for x in data)/len(data)
    std=(var)**0.5
    print(f"Mean : {mean}\n")
    print(f"Median : {median}\n")
    print(f"min : {minval}\n")
    print(f"max : {maxval}\n")
    print(f"var : {var}\n")
    print(f"std : {std}\n")
```

```python
for s in species:
    sp_data=group.get_group(s)
    
    for i in datatypes:
        print('\n',s,'-',i,'\n')
        statistic(sp_data[i])
```
