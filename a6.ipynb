```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset('iris')
```

```python
features=data.columns.tolist()
features
```

```python
dtyp=data.dtypes
dtyp
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,8))

sns.histplot(x='sepal_length',hue='species',palette='Set1',data=data,ax=axs[0,0])
sns.histplot(x='sepal_width',hue='species',palette='Set1',data=data,ax=axs[0,1])
sns.histplot(x='petal_length',hue='species',palette='Set1',data=data,ax=axs[1,0])
sns.histplot(x='petal_width',hue='species',palette='Set1',data=data,ax=axs[1,1])
plt.show()
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,8))

sns.boxplot(x='sepal_length',y='species',hue='species',palette='Set1',data=data,ax=axs[0,0])
sns.boxplot(x='sepal_width',y='species',hue='species',palette='Set1',data=data,ax=axs[0,1])
sns.boxplot(x='petal_length',y='species',hue='species',palette='Set1',data=data,ax=axs[1,0])
sns.boxplot(x='petal_width',y='species',hue='species',palette='Set1',data=data,ax=axs[1,1])
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(10,8))

sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',palette='Set1',data=data,ax=axs[0])

sns.scatterplot(x='petal_length',y='petal_width',hue='species',palette='Set1',data=data,ax=axs[1])

plt.show()
```
