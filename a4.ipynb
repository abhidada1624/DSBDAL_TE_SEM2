```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
data=sns.load_dataset('titanic')
```

```python
data
```

```python
plt.figure(figsize=(12,8))

sns.countplot(x='survived', hue='survived',palette='Set1',data=data)
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.countplot(x='pclass', hue='survived',palette='Set1',data=data)
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(x='pclass',y='age', hue='survived',palette='Set1',data=data)
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.countplot(x='pclass', hue='pclass',palette='Set1',data=data)
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.countplot(x='sex', hue='sex',palette='Set1',data=data)
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.histplot(x='fare',bins=20,kde=True,data=data)
plt.show()
```
